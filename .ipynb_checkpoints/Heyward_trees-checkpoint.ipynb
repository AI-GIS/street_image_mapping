```python
# put those overlaping masks into a group

import numpy as np

# Example masks (binary numpy arrays)
mask1 = np.array([[True, False, False], [False, True, False], [False, False, False]])
mask2 = np.array([[False, False, True], [False, True, False], [True, False, False]])
mask3 = np.array([[False, False, False], [False, False, False], [False, False, True]])
mask4 = np.array([[True, False, False], [True, True, False], [False, False, False]])

# List of masks
masks = [mask1, mask2, mask3, mask4]

# Initialize each mask in its own group
groups = [{i} for i in range(len(masks))]

# Function to find the group that a mask belongs to
def find_group(mask_index, groups):
    for group in groups:
        if mask_index in group:
            return group
    return None

# Check for overlaps and group masks together
for i in range(len(masks)):
    for j in range(i + 1, len(masks)):
        overlap = np.logical_and(masks[i], masks[j])
        if np.any(overlap):
            group_i = find_group(i, groups)
            group_j = find_group(j, groups)
            if group_i != group_j:
                # Merge the groups
                group_i.update(group_j)
                groups.remove(group_j)

# Output the groups of overlapping masks
grouped_masks = []
for group in groups:
    grouped_masks.append([masks[i] for i in group])

# Output the indices of the groups
for idx, group in enumerate(groups):
    print(f"Group {idx}: Masks with indices {list(group)}")
```

```python
def bounding_box(binary_array):
    # Find the indices of the non-zero elements
    rows, cols = np.where(binary_array)
    
    if len(rows) == 0 or len(cols) == 0:
        # If the array is empty or has no True elements
        print(rows, cols)
        return None
        # Calculate the bounding box
    left = np.min(cols)
    top = np.min(rows)
    right = np.max(cols)
    bottom = np.max(rows)
    
    return [left, top, right, bottom]
    
def get_mask_ratio(arr):
    mask_area = arr.sum()
    # print("arr shape:", arr.shape)
    try:
        left, top, right, bottom = bounding_box(arr)
        # print("left, top, right, bottom:", left, top, right, bottom)
    except:
        return -1
        # pass
    # 
    total_area = (bottom - top) * (right - left)
    ratio = mask_area / total_area
    return ratio
 


get_mask_ratio(detection[1].mask[1])
```

```python
import matplotlib.pyplot as plt


        # ax.plot(ratio_list)
     

def cut_tree_canopy(detection, trunk_area_threshold=0.7, interval_pix=20): 
    # not applicable when the tree root like a canopy (e.g., "L" shape tree), which will never reach the trunk_area_threshold
    mask_cnt = len(detection[1].mask)
    # print("mask cnt:", mask_cnt)
    fig, ax = plt.subplots()
    
    for i in range(mask_cnt):  
        if i == 0:
            continue
        ratio_list = []
        bottom = int(detection[1].xyxy[i][-1])
        is_cutoff = False
        for top in range(interval_pix, bottom, interval_pix):
            
            ratio = get_mask_ratio(detection[1].mask[i][top:])
            if ratio == -1:
                break
            # print(top, bottom, ratio)
            ratio_list.append(ratio)
            if ratio > trunk_area_threshold and (not is_cutoff):                
                detection[1].mask[i][:top, :] = False
                bbox = bounding_box(detection[1].mask[i])
                detection[1].xyxy[i] = bbox
                is_cutoff = True
                # print("cut. ", trunk_area_threshold)
                # break
            # print(detection[1].xyxy[i])
        
        ax.plot(ratio_list)
        # break

        
def is_stick(arr, cut_1=0.25, cut_2=0.75, dif_thres=0.2):
    '''
    A stick, like tree trunk, its widths at cut_1 and cut_2 points should have a difference < dif_thres
    '''
    height = arr.shape[0]
    cut_1_row = int(height * cut_1)
    cut_2_row = int(height * cut_2)
    
    # find the width
    cols = np.where(arr[cut_1_row])
    if len(cols) == 0:
        return False
    cut_1_width = cols[-1] -  cols[0] 
    print("cut_1_width:", cut_1_width)
    print("cols.shape:", cols.shape)
    if cut_1_width.shape[0] == 0:
        return False
    
    print("cols:", cols, len(cols))
    
    print("cut_1_width:", cut_1_width)
    
    cols = np.where(arr[cut_2_row])
    if len(cols) < 2:
        return False
    cut_2_width = cols[-1] -  cols[0] 
    if cut_2_width < 2:
        return False
    
    dif = (cut_2_width - cut_1_width) / min(cut_2_width, cut_1_width)
    dif = abs(dif)
    
    if dif > 0.2:
        return False
    if dif <= 0.2:
        return True
    

def keep_trunk(detection, cut_1=0.25, cut_2=0.75, dif_thres=0.2, interval_pix=10):    
    mask_cnt = len(detection[1].mask)
    # print("mask cnt:", mask_cnt)
    fig, ax = plt.subplots()
    
    for i in range(mask_cnt):  
        if i == 0:
            continue
        dif_list = []
        bottom = int(detection[1].xyxy[i][-1])
        is_cutoff = False
        for top in range(interval_pix, bottom, interval_pix):
            
             if is_stick(detection[1].mask[i][top:], cut_1, cut_2, dif_thres):           
                detection[1].mask[i][:top, :] = False
                bbox = bounding_box(detection[1].mask[i])
                detection[1].xyxy[i] = bbox
                is_cutoff = True

                
# cut_tree_canopy(detection, trunk_area_threshold=trunk_area_threshold, interval_pix=20)
# plt.show()

keep_trunk(detection, cut_1=0.25, cut_2=0.75, dif_thres=0.2, interval_pix=10)
```
