```python

# plt.axis('scaled')
fig, ax = plt.subplots()

ax.plot([pano1.x, pano2.x], [pano1.y, pano2.y])
ax.scatter([pano1.x, pano2.x], [pano1.y, pano2.y])
ax.axis('equal')

# Create an arrow patch
# Convert angle to radians

def draw_yaw_arrow(ax, pano, length=3, width=0.3):
    '''
    length: 
    '''
    length = 3
    
    yaw = np.radians(90 - pano.jdata['Projection']['pano_yaw_deg'])    

    # Draw pano yaw
    x_end =   length * np.cos(yaw)
    y_end =   length * np.sin(yaw)
    arrow = patches.Arrow(x=pano.x, y=pano.y, dx=x_end, dy=y_end, width=0.6, color="blue")
    ax.add_patch(arrow)
    
    ax.text(x=pano.x + 1, y=pano.y + 1, s=pano.panoId, c='blue')
    
    
    
    length = 20
    yaw = np.radians(90 - pano.trunk_heading_deg)    

    # draw trunk heading
    x_end =   length * np.cos(yaw)
    y_end =   length * np.sin(yaw)
    arrow = patches.Arrow(x=pano.x, y=pano.y, dx=x_end, dy=y_end, width=0.6, color="g")
    ax.add_patch(arrow)
    
    ax.set_ylim(min(pano.y, y_end + pano.y) - 10, max(pano.y, y_end + pano.y) + 10)
    ax.set_xlim(min(pano.x, x_end + pano.x) - 10, max(pano.x, x_end + pano.x) + 10)

    heading_dif_deg = utils.degree_difference(pano.jdata['Projection']['pano_yaw_deg'], pano.trunk_heading_deg)
    print(pano.panoId, "heading_dif_deg:", heading_dif_deg)

for pano in [pano1, pano2]:
    draw_yaw_arrow(ax, pano)
# Add the patch to the axes

plt.show()
```

```python
def two_points_distance(x1, y1, x2, y2):
    distance = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    # print("pano1.x, pano1.y, pano2.x, pano2.y: ", pano1.x, pano1.y, pano2.x, pano2.y)
    # print(distance)
    return distance

def two_panos_distance(pano1, pano2):  # not finished
    if pano1.x is not None:
        distance = two_points_distance(pano1.x, pano1.y, pano2.x, pano2.y)
        print("pano1.x, pano1.y, pano2.x, pano2.y: ", pano1.x, pano1.y, pano2.x, pano2.y)
        print(distance)

    else:
        print("Need to compute pano1.x/y, pano2.x/y first!")
        return None

    return distance

distance = utils.two_panos_distance(pano1, pano2)

angle_3_rad = math.radians(angle_3)
D = distance / math.sin(angle_3_rad)

dis_1 = D * math.sin(math.radians(angle_1))
dis_2 = D * math.sin(math.radians(angle_2))


distance, D
dis_1, dis_2
```

```python
import math

def row_col_to_angle(row, col, width, height, horizontal_fov_rad):
  
    # Normalize pixel coordinates
    x = col - width / 2
    y = height / 2 - row
    
    
    azimuth_max = horizontal_fov_rad / 2 
    R = (width / 2) / math.tan(azimuth_max)  # in pixel

    azimuth_rad = math.atan(x / R)    # = math.atan(x / R) 
    
    hypotenuse = R / math.cos(azimuth_rad)
    
    altitude_rad = math.atan(y / hypotenuse)
    
    print("x, y, R:", x, y, R)
    print("azimuth_rad, math.cos(azimuth_rad):", azimuth_rad, math.cos(azimuth_rad))
 
    return azimuth_rad, altitude_rad
 
 
# row, col = 419, 565  # Center pixel

row, col = 768/2, 1024/2   
# row, col = 0, 0  
row, col = 768, 512  


width = 1024
height = 768

horizontal_fov_rad = math.radians(45)
vertical_fov_rad = math.radians(34.5158770189817)


azimuth_rad, altitude_rad = row_col_to_angle(row, col, width, height, horizontal_fov_rad)

altitude_deg = math.degrees(altitude_rad)
azimuth_deg = math.degrees(azimuth_rad)

print(f"altitude_deg: {altitude_deg}, azimuth_deg: {azimuth_deg}")
print(azimuth_deg)
```
